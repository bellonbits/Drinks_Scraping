```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# Set up Selenium WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# URL to scrape
url = "https://www.jumia.co.ke/mlp-black-friday/groceries/#catalog-listing"
driver.get(url)

# Give the page some time to load
time.sleep(5)

# Get the page source and parse with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all product elements
products = soup.find_all('div', class_='info')

# Loop through products and extract data
for product in products:
    # Get product name
    name = product.find('h3', class_='name').text if product.find('h3', class_='name') else 'No name'
    
    # Get current price
    current_price = product.find('div', class_='prc').text if product.find('div', class_='prc') else 'No price'
    
    # Get old price (if available)
    old_price = product.find('div', class_='old').text if product.find('div', class_='old') else 'No old price'
    
    # Get discount (if available)
    discount = product.find('div', class_='bdg _dsct _sm').text if product.find('div', class_='bdg _dsct _sm') else 'No discount'
    
    # Get ratings (if available)
    ratings = product.find('div', class_='stars _s').text if product.find('div', class_='stars _s') else 'No ratings'
    
    # Print the scraped data
    print(f"Name: {name}\nCurrent Price: {current_price}\nOld Price: {old_price}\nDiscount: {discount}\nRatings: {ratings}\n")

# Close the WebDriver
driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
import time

# Set up Selenium WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Base URL to scrape
base_url = "https://www.jumia.co.ke/mlp-black-friday/drinks/#catalog-listing"

# Initialize an empty list to store product data
product_data = []

# Function to extract data from the current page
def extract_data(page_source):
    soup = BeautifulSoup(page_source, 'html.parser')
    products = soup.find_all('div', class_='info')
    
    for product in products:
        # Get product name
        name = product.find('h3', class_='name').text.strip() if product.find('h3', class_='name') else 'No name'
        
        # Get current price
        current_price = product.find('div', class_='prc').text.strip() if product.find('div', class_='prc') else 'No price'
        
        # Get old price (if available)
        old_price = product.find('div', class_='old').text.strip() if product.find('div', class_='old') else 'No old price'
        
        # Get discount (if available)
        discount = product.find('div', class_='bdg _dsct _sm').text.strip() if product.find('div', class_='bdg _dsct _sm') else 'No discount'
        
        # Get ratings (if available)
        ratings = product.find('div', class_='stars _s').text.strip() if product.find('div', class_='stars _s') else 'No ratings'
        
        # Append the product details to the list
        product_data.append({
            'Name': name,
            'Current Price': current_price,
            'Old Price': old_price,
            'Discount': discount,
            'Ratings': ratings
        })

# Function to scrape Jumia across all pages
def scrape_jumia():
    driver.get(base_url)
    time.sleep(5)  # Wait for the page to load
    
    while True:
        # Extract data from the current page
        extract_data(driver.page_source)
        
        try:
            # Find the 'Next' button using its aria-label and click it
            next_button = driver.find_element(By.CSS_SELECTOR, "a.pg[aria-label='Next Page']")
            ActionChains(driver).move_to_element(next_button).click().perform()
            
            # Wait for the next page to load
            time.sleep(5)
        except Exception as e:
            print(f"No more pages or error occurred: {e}")
            break

# Start scraping
scrape_jumia()

# Close the WebDriver
driver.quit()

# Convert the product data into a DataFrame
df = pd.DataFrame(product_data)

# Display the first few rows of the DataFrame
print(df.head())

# Optionally, save the DataFrame to a CSV file
df.to_csv('jumia_products.csv', index=False)
```
